```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

```python
df_housing = pd.read_csv('Housing.csv')
```

```python
df_housing.head()
```

```python
df_housing.shape
```

```python
#Converting columns into category
for col in df_housing.columns:
    if df_housing[col].dtype == "object":
        df_housing[col] = df_housing[col].astype("category").cat.codes
```

```python
df_housing.corr()
```

```python
#Graph showing the correlation using matplotlib and seaborn - verify which non categorical value can be used for the model application
plt.figure(figsize=(10, 8))
sns.heatmap(df_housing.corr(), annot=True, linewidth=.5)
plt.title('Correlation')
plt.show()
```

```python
# Correlation value based on the graph results, which most part of variables have the value based on 0.4 - the variable for model aplication is 'area'
corr_threshold = 0.4

features = [
    df_housing.columns[idx] for idx, corr in enumerate(df_housing.corr()["price"]) if abs(corr) > corr_threshold and df_housing.columns[idx] != "price"
]
```

```python
# Model forlinear regression analysis according to each feature in the dataset
for feature in features:

    model = LinearRegression()

    X = df_housing[[feature]]
    y = df_housing["price"]

    model.fit(X, y)
    print("-----------------------------------------------------------")
    plt.scatter(X, y, color='blue', label='Actual Price')
    plt.plot(X, model.predict(X), color='red', label='Predicted Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    X = sm.add_constant(X)
    print("Este modelo é: Preço = {:.5} + {:.5}X".format(model.intercept_, model.coef_[0]))
    model = sm.OLS(y, X).fit()
    print(model.summary())
    print("-----------------------------------------------------------")
```
